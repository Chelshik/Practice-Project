## Loading And Exploring Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Loading data directly form the raw link
df=pd.read_csv("https://raw.githubusercontent.com/FlipRoboTechnologies/ML-Datasets/main/Grades/Grades.csv")
```

```python
# checking out the dataset
df
```

This dataset contain 571 rows and 43 columns . Number of features is huge so later with the help of PCA. Most of the feaures are categorical and our target i.e CGPA continuous values and it is a regression problem.

```python
# Checking for missing values 
df.info()
```

All the features are of object datatype and our target variable is of float datatype.

```python
df.isnull().sum()
```

As we can see there are many missing values present in our dataset.

```python
# Checking number of unique values in each column
df.nunique()
```

```python
df= df.drop(columns=['Seat No.'])
```

```python
df.sample(3)
```

### Handling Missing Values

```python
categorical_columns = df.select_dtypes(include=['object']).columns

# Filling missing values in categorical columns with the mode
for column in categorical_columns:
    df[column] = df[column].fillna(df[column].mode()[0])

# Checking the DataFrame after handling missing values
print(df.info())
```

### Encoding Categorical Columns

```python
# Converting categorical columns into numerical columns
from sklearn.preprocessing import LabelEncoder

# Initializing label encoder
label_encoder = LabelEncoder()

# Applying label encoding to each categorical column
for column in df.select_dtypes(include=['object']).columns:
    df[column] = label_encoder.fit_transform(df[column])

print(df)
```

### Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
x=df.iloc[:,0:-1]
x.shape
```

```python
pca=PCA(n_components=9)
```

```python
y=df.iloc[:,-1]
```

```python
x=pca.fit_transform(x)
```

```python
x.shape
```

```python
x
```

```python
# Calculating the number of principal components
num_components = x.shape[1]

# Creating subplots for visualizations
fig, axes = plt.subplots(nrows=num_components, ncols=num_components, figsize=(15, 15))

# Looping through each pair of principal components
for i in range(num_components):
    for j in range(num_components):
        if i != j:
            axes[i,j].scatter(x[:, i], x[:, j], c=y, cmap='viridis', alpha=0.5)
            axes[i,j].set_xlabel(f'PC {i+1}')
            axes[i,j].set_ylabel(f'PC {j+1}')
        else:
            axes[i,j].axis('off') 

# Adding a title to the entire plot
plt.suptitle('Pairwise Visualization of Principal Components')

# Adjusting layout to prevent overlap of subplots
plt.tight_layout()

plt.show()
```

```python
# Calculating the correlation matrix
correlation_matrix = np.corrcoef(x, rowvar=False)

# Creating a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, xticklabels=np.arange(1, x.shape[1]+1), yticklabels=np.arange(1, x.shape[1]+1))
plt.title('Correlation Heatmap of Principal Components')
plt.xlabel('Principal Components')
plt.ylabel('Principal Components')
plt.show()
```

## Train Test Split

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Initializing and training the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting on the test set
y_pred = model.predict(X_test)

# Evaluating the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```

## Saving the Model
